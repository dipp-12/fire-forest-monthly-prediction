```python
import pickle
import pandas as pd
from tqdm import tqdm
```

# Load Pickle Data

```python
with open('/kaggle/input/west-kalimantan-mod21c3/data/raw_data_mod21c3.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
list = []
coordinate = [str(item[1:3]) for item in data[0]]

for day in tqdm(range(len(data))):
    daily_data = data[day]
    date = daily_data[1][0]
    year = date.split("_")[0]
    temp_list = [item[-1] for item in daily_data]
    temp_list.insert(0, year)
    temp_list.insert(1, date)
    list.append(temp_list)
    
df = pd.DataFrame(list, columns=['year', 'date'] + coordinate)
```

# Remove Label Column

```python
df.drop(df.columns[2], axis=1, inplace=True)
```

# Save CSV

```python
df.to_csv('/kaggle/working/data_mod21c3.csv', index=False)
```
