### Drop columns with all null values

```python
df = df.dropna(axis=1, how='all')
df.head()
# Alternatively, fill with nearest value (nearest column)
```

### Fill null values "in between" with linear interpolation

```python
df = df.interpolate(method='linear') 
```

```python
df.isna().sum().sum()
```

```python
# df = df.dropna(axis=1, how='any')
```

### Fill null values on the first few row with the next value

```python
# Fill null value on the first few row with the next value
df = df.bfill()
```

```python
df.describe()
```

### Make sure no null values

```python
df.isna().sum().sum()
```

### Split data into training-validation, test, <s>and prediction</s>
1. Training-validation: 2019-2022
2. Test: 2023
3. <strike>prediction: 2024 or last 3 months</strike> (data taken separated on new file)

```python
group_by_year = df.groupby('year')
df_training_validation = pd.concat([group_by_year.get_group(year) for year in range (2019, 2023)])
df_test = group_by_year.get_group(2023)
# TODO: add data for prediction (2024 or last 3 months)
# df_predict = df.tail(3)
```

```python
df_training_validation.to_csv('data/training_validation_mod21c3.csv', index=False)
df_test.to_csv('data/test_mod21c3.csv', index=False)
# df_predict.to_csv('data/predict_mod21c3.csv', index=False)
```
